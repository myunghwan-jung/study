# AWS Support Plan
- AWS의 지원 정책
- 총 5가지 티어
    - 기본
    - 개발자
    - 비지니스
    - 엔터프라이즈 On-Ramp
    - 엔터프라이즈
- 티어 별로 제공 서비스와 비용이 다름

## 기본 서포트 플랜 혜택
- 모든 고객에게 제공
- 제공 서비스
    - 고객 서비스 및 커뮤니티
    - AWS Trusted Advisor 핵심 체크
        - EBS, RDS 퍼블릭 스냅샷
        - S3 버킷 권한
        - IAM 사용
        - 루트 어카운트에 MFA 설정 여부
        - 보안 그룹의 모든 포트가 허용되어 있는지 여부
- 기술 서포트 없음

## 개발자 서포트 플랜 혜택
- 기술 지원
    - 클라우드 지원 담당자에게 업무 시간에 이메일을 통한 액세스
- 응답 시간
    - 일반 : 24시간 이내
    - 시스템 손상 : 12시간 이내
- 비용
    - 월 $29 혹은 요금의 3% (둘 중 높은 것)

## 비지니스 서포트 플랜 혜택
- 기술 지원
    - 클라우드 지원 담당자에게 연중무휴 24시간 전화/이메일/채팅 액세스
- 응답 시간
    - 일반 : 24시간 이내
    - 시스템 손상 : 12시간 이내
    - 프로덕션 시스템 손상 : 4시간 이내
    - 프로덕션 시스템 중단 : 1시간 이내
- AWS Support API를 사용 가능한 최소 티어
- 추가 비용을 지불하고 인프라 이벤트 관리를 액세스
- 비용
    - 월 $100 혹은 요금의 10~3%(둘 중 높은 것)

## 엔터프라이즈 서포트 플랜 혜택
- 기술 지원
    - 클라우드 지원 담당자에게 연중무휴 24시간 전화/이메일/채팅 액세스
- 응답 시간
    - 일반 : 24시간 이내
    - 시스템 손상 : 12시간 이내
    - 프로덕션 시스템 손상 : 4시간 이내
    - 프로덕션 시스템 중단 : 1시간 이내
    - 비즈니스 크리티컬 시스템 중단 : <30분(Ramp), 15(Enterprise)
- 비용
    - 월 $5,500~15,000 혹은 월 사용료의 10~3%
- 전담 테크니컬 어카운트 관리자(TAM)
    - AWS에서 지정된 기술 매니저
    - 매번 다른 사람이 배정되는 것이 아닌, 나의 어카운트만을 위한 전담 배정된 기술 매니저
- 컨시어지 지원 팀
    - 엔터프라이즈 플랜 이상에서만 배정되는 계정 지원 전담
- 인프라 이벤트 관리
    - 별도의 비용 없이 인프라 이벤트 관리(비즈니스 플랜의 경우 추가 요금 필요)

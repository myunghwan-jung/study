# CloudWatch 기초
- CloudWatch는 Devops, 개발자, SRE 및 IT 관리자를 위해 구축된 모니터링 및 관찰 기능 서비스입니다. 클라우드워치는 어플리케이션을 모니터링하고, 시스템 전반의 성능 변경 사항에 대응하며, 리소스 사용률을 최적화하고, 운영 상태에 대한 통합된 보기를 확보하는데 필요한 데이터와 실행 가능한 통찰력을 제공합니다.
- AWS에서 제공하는 서비스/어플리케이션 모니터링 서비스
- Public 서비스
    - 인터넷을 통해서 접근
    - 혹은 인터페이스 엔드포인트로 접근
- 로그, 지표, 이벤트 등의 운영데이터를 수집하여 시각화 및 처리
    - 경보 생성을 통해 자동화 대응 가능
- AWS의 거의 대부분의 서비스와 연동 가능

## 주요 기능
- 지표 수집
    - 시간 순서로 정리된 데이터의 집합
        - 다수의 데이터 포인트로 구성
    - AWS의 대부분의 서비스는 기본적으로 지표를 제공
        - 예 : EC2 CPU 사용량, 네트워크 등
    - 커스텀 지표 생성 가능
        - 유저가 직접 원하는 데이터포인트를 생성해서 클라우드워치로 전달하여 생성
- 알람
    - 수집된 지표 값의 변동에 따라 발생하는 알림 생성
        - 일정 수치로 도달하거나 이상/이하일 때 이벤트 발생
    - 다양한 방법으로 대응 가능
        - SNS로 Lambda 실행, 이메일 전달 등
        - 예 : 웹서버의 500에러가 일정 수치 이상일 때 슬랙 알림
- 로그 수집 및 관리
    - 람다, EC2, Route53, ECS등 여러 AWS 서비스의 로그를 수집
    - 수집된 로그를 Kinesis, S3등 다른 서비스/게정으로 전달 가능
    - 혹은 자체적으로 확인하거나 쿼리 가능
- 대시보드
    - 수집한 로그/지표를 기반으로 대시보드 구성
    - 외부 리소스를 활용해서 커스텀 대시보드 구성 가능
        - 예 : S3 객체 표시. HTML 커스텀 그래프 표시
- 기타 서비스
    - 어플리케이션을 모니터링하는 Synthetics Canary
    - 어플리케이션, 컨테이너, 람다 등의 문제를 찾아주는 인사이트 서비스 등

# 클라우드워치 지표 수집/알람 데모

## 커스텀 지표 받기(EC2 메모리 모니터링)
- EC2에 웹서버 설치
- 클라우드워치 에이전트 설치 및 설정
    - 로그 수집/커스텀 지표(메모리/디스크 수집) 용도
- 이후 커스컴 지표 확인
- 수집된 Access Log를 지표 필터로 지표화
- 수집된 지표 기반으로 알람 설정 : 404 에러가 분당 5번 이상 발생 시 메일 전송

## 실습 순서
1. IAM 역할 생성
    - 사용자 : EC2
    - cloudwatchfullaccess v2 권한 부여
    - ec2readonlyaccess 권한 부여
    - demo-ec2-role-for-cloudwatch
2. EC2 인스턴스 생성
    - demo-cloudwatch
    - 보안그룹 디폴트
    - 고급 세부 정보
        - IAM 프로파일 적용
3. 인스턴스 연결
    - 스크립트 명령어 입력(httpd.conf) : 웹서버와 클라우드워치 에이전트 설치 및 설정
4. 인스턴스의 퍼블릭 아이피로 접속, 새로고침으로 로그를 남기자
5. 클라우드 워치로 이동
    - 로그 그룹 조회 : 인스턴스 아이디로 기록된 로그스트림 조회. 기타 여러 로그들도 조회해보자
    - 모든 지표 : CWAgent에 들어가면 메모리 로그 조회 가능
        - 그래프로 표시된 지표에서 기간, 간격 등을 조절해보자
    - 인스턴스 아이디로 지표를 조회하면 AWS에서 자체적으로 수집하는 지표들을 조회할 수 있다
6. 알람 설정을 해보자
    - apache/access 로그 그룹으로 이동, 작업 - 지표 필터 생성, 패턴에 스크립트 넣자
    - 패턴 테스트
    - 필터 이름, 지표 네임스페이스 demo-404
    - 지표 이름 demo-404-metric
    - 지표 값 : 1
    - 인스턴스의 퍼블릭 ip에 접속해서 에러를 발생시켜보자
    - 모든 지표에서 demo-404 조회 (1시간, 1분, 합계)
    - 경보 : 모든 경보, 지표 선택
        - demo-404에서 지표 선택
        - 통계 : 합계
        - 기간 : 1분
        - 보다 큼, 5개
        - 새 주제 생성, 이메일 입력
        - 주제 생성, 메일 확인 : confirm subscription
        - 다시 에러를 생성해보자
7. 리소스 정리 : 경보, 인스턴스 삭제, 로그 그룹 삭제(apache)

# 예약 인스턴스

## EC2 요금 모델
- on-demand : 사용한 시간 만큼 요금 지불
- Spot Instance : 남는 인스턴스를 저렴하게 사용
- Reserved Instances : 인스턴스 사용 기간을 약정
- Dedicated : 물리적인 전용 인스턴스 임대
    - Dedicated Instance / Dedicated Host
- Saving Plan : AWS의 컴퓨팅 사용량을 약정

### On-Demand
- 실행하는 인스턴스에 따라 초당 혹은 시간당 컴퓨팅 파워로 측정된 가격을 지불
- 약정 필요 없음
- 수요 예측이 힘들거나 유연하게 EC2를 사용하고 싶을 때
- 한번 써보고 싶을 때

### Reserved Instances
- EC2 인스턴스를 일정 기간 약정하여 요금을 할인받는 방식
    - 온디맨드 EC2 사용 요금을 할인 받는 방식으로 적용
    - 할인 받고 싶은 EC2 인스턴스와 같은 리전, 같은 유형 구매 필요
- 약정 기간이 길 수록 더 큰 할인율 적용
    - 1년 혹은 3년 선택 가능
- 리전 별로 적용 : 다른 리전과 공유 불가

### Reserved Instances 구성 요소
- 인스턴스 유형 : 타입, 크기
- 테넌시 : 전용 호스트 혹은 공유 하드웨어
- 플랫폼 : Windows or Linux
- 기간 약정 : 1년 or 3년
- 결제 방법
    - 전체 선결제 : 모든 금액을 기간 시작 전에 결제 : 가장 저렴
    - 부분 선결제 : 비용 중 일부만 시작 전에 결제, 나머지 비용은 할인 가격으로 시간 당 청구
    - 선결제 없음 : 모든 비용을 할인 가격으로 시간당 청구
- 제공 클래스
    - 표준 : 큰 할인 혜택, 단 교환 불가능 (수정만 가능)
    - 컨버터블 : 낮은 할인 혜택. 다른 속성의 예약 인스턴스로 교환 가능

### 예약 인스턴스 종류
- 범위에 따라
    - 리전 : 리전 전체에 사용할 수 있는 예약 인스턴스
        - 미리 용량 예약 불가능 : 언제나 원할 때 사용할 수 있는 것이 아니다
        - 인스턴스 크기 유연성 적용 : 크기에 상관 없이 인스턴스 패밀리의 사용량으로 예약 인스턴스 소모
        - 구매 대기 가능
    - 영역 : 특정 가용영역에서만 사용할 수 있는 예약 인스턴스
        - 미리 용량 예약 가능 : 따로 해당 유형의 인스턴스를 확보
        - 구매 대기 불가능
- 제공 클래스에 따라
    - 표준
        - 교환 불가능
        - 예약 인스턴스 마켓플레이스에 판매/구매 가능
    - 전환형
        - 인스턴스 유형, 플랫폼, 범위 또는 테넌시 등의 다른 속성의 전환형 예약 인스턴스와 교환 가능
        - 예약 인스턴스 마켓플레이스에 판매/구매 불가능

### 예약 인스턴스 적용 방식
- 구매 즉시 할인 가격 적용
- 정확하게 설정한 속성과 맞아야 함
    - 인스턴스 유형, 플랫폼, 가용영역. 인스턴스 사이즈 등
- 인스턴스 사이즈 유연성(리전 예약 인스턴스 only)
    - 각 인스턴스 사이즈 별로 사용 점수를 보유
    - 구매한 RI 포인트로 사용중인 인스턴스 포인트를 커버, 남는 부분은 온디맨드 과금
    - 지원하지 않는 OS/인스턴스 타입 존재
        - 지원하지 않는 OS : Windows Server(SQL), RHEL, SUSE Linux Enterprise Server 등
        - 지원하지 않는 인스턴스 타입 : G4ad, G4dn, G5, G5g, Inf1, Inf2 등

### 예약 인스턴스 적용
- 매시 정각에 비용 발생(선결제 없는 경우)
    - 오픈 소스 리눅스의 경우 할인 혜택 자체는 초 단위로 적용
    - 상용 리눅스의 경우 할인 혜택이 시간 단위로 적용
- 1 RI당 한 시간의 인스턴스만 커버
    - m4.xlarge 1 RI를 구매 후 m4.xlarge 4개를 1시간 동안 구동하는 경우 : 1개만 커버되고 나머지 3개는 온디맨드 요금 발생
    - m4.xlarge 1 RI를 구매 후 m4.xlarge 4개를 15분 동안 구동하는 경우 : 모두 커버

### 전환형 예약 인스턴스의 교환
- 전환형 RI의 경우 교환 가능
    - 규칙에 맞는 교환이라면 횟수 제한 없음
    - 조건 : 내가 보유한 RI 가치가 받으려는 RI 가치보다 작거나 같은 경우

### 전환형 인스턴스의 교환 규칙
- 현재 AWS에서 제공하는 RI 타입으로만 교환 가능
- 다른 리전이랑 교환 불가능
- 예약 인스턴스를 한개씩만 교환 가능 (받을 때는 여러 인스턴스로 받을 수 있음)
- 일부만 교환하고 싶을 떄는 예약 인스턴스를 작은 단위로 나눈 후 각각의 단위를 교환
- 약정 기간이 같은 예약 인스턴스만 교환 가능(1년, 3년)

### 표준 예약 인스턴스 수정
- 가용 영역, 인스턴스 사이즈(같은 유형), 범위 변경 가능
- 분할 가능
    - 1 x t2.small => 2 x t2.micro
    - 10개 AZ a RI => 5개 AZ a, 5개 AZ b
- 병합 가능
    - 2 x t2.micro => 1 x t2.small
- 인스턴스 사이즈 변경은 리눅스/유닉스만 가능
- 몇 가지 인스턴스 유형은 불가능 (G4)

### 예약 인스턴스 구매/판매
- AWS 마켓플레이스에서 다른 사람이 판매하는 예약 인스턴스 구매 가능
- 구매한 예약 인스턴스가 필요 없어지면 판매도 가능
- 표준 예약 인스턴스만 판매, 구매 가능
- 판매 선결제 금액의 12%를 AWS에서 수수료로 가져감
- 미국 은행 계좌가 있어야 판매 가능, 한국에서 사용하기 어려움
- 경우에 따라서는 MSP에 판매 가능

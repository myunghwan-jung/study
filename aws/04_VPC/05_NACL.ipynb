# NACL (Network Access Control List)
- NACL은 1개 이상의 서넷 내부와 외부의 트래픽을 제어하기 위한 방화벽 역할을 하는 VPC를 위한 선택적 보안 계층입니다.
- 보안 그룹처럼 방화벽 역할을 담당
- 서브넷 단위
    - 인스턴스 단위로 제어 불가능
    - 다양한 서브넷에 연동 가능
- Port , IP를 직접 deny 가능
    - 외부 공격을 받는 상황 등 특정 IP를 블록하고 싶을 때 사용
- Stateless
    - 들어오는 트래픽과 나가는 트래픽을 구분, 서로 다른 규칙이 적용된다.
    - 즉 일반적으로 Outbound에 Ephemeral Port 범위를 열어주어야 정상적인 통신이 가능

## NACL 규칙
- 규칙 번호 : 규칙에 부여되는 고유 숫자이며, 낮은 번호부터 규칙이 평가된다
    - AWS 추천은 100 단위 증가
- 유형 : 트래픽 유형(SSH=22, DNS=53, UDP=17 등)
- 프로토콜 : 통신 프로토콜 (예 : TCP, UDP, SMP 등)
- 포트 범위 : 허용 혹은 거부할 포트 범위
- 소스 : IP 주소의 CIDR 블록
- 허용 or 거부
- 예시 :
    - 112.12.35.4 의 해킹을 차단하고 싶다면?
    - 규칙번호 100에서 모든 소스를 허용, 규칙변호 200에서 112.12.35.4 를 차단하면?
    - 규칙번호 100이 먼저 평가를 받고, 112.12.35.4를 허용해버려서 규칙번호 200은 의미가 없어짐.

## NACL 고려할 점
- Stateless 방화벽이기 때문에 원활한 통신을 위해서는 아웃바운드도 신경써야 함
    - 임시포트 Allow!
        - Linux : 32768~61000
        - Windows XP : 1025~5000, Windows vista~ : 49152~65535
        - 따라서 1025~65535 까지 열어두는 것이 일반적
- 서브넷에서 나가거나 들어오는 트래픽에만 적용
    - 즉 서브넷 내부 트래픽에는 적용되지 않음
- 다양한 서브넷으로 구성되어있는 멀티 티어 아키텍처라면 더 많은 고민이 필요
    - 따라서 가능한 보안그룹으로 처리하고, 꼭 필요할 때 NACL을 사용하자
- VPC 생성 시, 혹은 AWS 계정 생성 시 주어지는 VPC에 기본 하나 제공
    - 자동 생성되는 NACL은 모든 트래픽 Allow
    - 직접 생성하는 NACL은 모든 트래픽 Deny
- 하나의 서브넷은 하나의 NACL만 연동 가능, 하지만 하나의 NACL은 여러 서브넷에 연동 가능.

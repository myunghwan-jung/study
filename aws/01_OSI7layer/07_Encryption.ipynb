# 암호화(Encryption) 및 SSL/TLS
- 암호화 : 수학적인 과정으로 어떤 정보를 읽을 수 없는 데이터로 변환하는 행위
    - 복원이 가능함 <-> Hash는 복원이 불가능
- 암호화의 종류
    - 암호화 기술을 사용하는 상황에 따라
        - 저장된 데이터 보호 (Encryption At Rest)
        - 전송 중 데이터 보호 (Encryption At Transit)
    - 암호화 방법에 따라
        - 대칭키 암호화
        - 비대칭키 암호화

## 암호화 용어
- 평문 : 암호화 하기 전의 데이터 (Planetext)
- 암호문 : 평문을 암호 키와 알고리즘을 사용해 암호화 한 데이터 (Ciphertext)
- 암호화 : 평문에 암호화 알고리즘과 키를 적용하여 키를 소유한 주체가 아니면 알아볼 수 없는 암호문으로 만드는 과정 (Encryption)
- 복호화 : 암호문에 키와 복호화 알고리즘을 적용하여 평문으로 되돌리는 과정 (Decryption)
- 키 : 평문을 암호화하거나 암호문을 평문으로 돌리기 위한 알고리즘의 핵심 가변정보 값
- 암호 알고리즘 : 암호화/복호화를 위해 사용되는 알고리즘 : AES, DES등.

### 저장된 데이터 보호
- 어플리케이션에서 저장 공간에 데이터를 저장할 때 키를가지고 암호화, 꺼내올 때 복호화.
- 저장 공간이 도난당해도 키가 없으니 복호화 불가능.
- 주로 하나의 물리적인 기기에 보안을 적용하기 위해 사용

### 전송 중 데이터 보호
- 서로 다른 서버끼리 데이터를 주고 받을 때 적용
- 주로 여러 시스템/기기 간 보안을 적용하기 위해 사용
    - 예 : SSH/TLS, HTTPS 등

### 대칭키 암호화
- 하나의 키로 암호화, 복호화를 하는 알고리즘
- 연산이 빠름
- 키를 전달하는 방법에 대한 고민이 필요.

### 비대칭키 암호화
- 한 쌍의 키를 활용한 암호화, 단 하나의 암호는 암호화만 가능하며, 다른 하나는 복호화만 가능
- 연산이 비교적 복잡하지만 키 전달이 쉬움
- 공개 키 / 비밀 키 : 각각 암호화, 복호화를 할 수 있는 키쌍
    - 예 : 공개 키로 암호화하면 비밀 키로 복호화
    - 예 : 비밀 키로 암호화하면 공개 키로 복호화

### 암호화 서명
- 키를 사용해서 데이터의 생성자가 데이터를 생성했음을 보장하는 방법
    - Private 키를 사용하여 Puclic 키로 검증 가능한 데이터의 서명을 생성
    - Private 키를 가진 주체가 데이터를 생성했음을 보장할 수 있는 방법

## SSL/TLS
- 클라이언트와 서버간에 데이터의 무결성과 기밀성을 보장할 수 있는 프로토콜
    - 상호간의 통신은 암호화되어 전달되며, 중간에 데이터를 탈취당해도 안전함
- HTTPS등 다양한 프로토콜에 활용
    - 즉 일반 HTTP의 경우 연결이 안전하지 않기 때문에 보안적으로 취약함
- 3단계 과정을 거침
    1) Cipher Suites 교환
    - Cipher Suites : TLS에서 활용하는 보안 알고리즘들
    - 클라이언트는 서버로 SSL/TLS 버전, 지원하는 Cipher Suites 목록, 기타 정보를 전송
    - 서버는 클라이언트로 SSL/TLS 버전, 선택한 Cipher Suites, 인증서, 기타 정보를 전송
    2) 인증
    - 인증 : 믿을 수 있는 기관(Certificate Authority)에서 배포하는 서버 신원에 대한 검증 확인
    - 인증서와 CA 키 활용
    3) 키 교환

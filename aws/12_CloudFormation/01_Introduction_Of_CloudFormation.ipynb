# AWS CloudFormation 소개
- AWS CloudFormation은 AWS 리소스를 모델링하고 설정하여 리소스 관리 시간으 줄이고 AWS에서 실행되는 어플리케이션에 더 많은 시간을 할애할 수 있도록 해주는 서비스입니다
- 인프라를 코드로 구성할 수 있는 무료 서비스(Infrastructure as Code)
- 언제 사용할까?
    - 인프라를 자동으로 프로비전하고 싶을 때 사용
    - 인프라를 재사용하거나 공유하고 싶을 때 사용
    - 정형화된 아키텍처를 코드화시켜 관리하고 싶을 때
    - 인프라의 프로비전 및 업데이트 과정을 관리하고 싶을 때
- 어떻게 사용할까?
    - 코드 기반으로 AWS 인프라를 여러 리전에 자동으로 프로비전하거나 업데이트
        - 이 과정에서 오류 발생 시 자동으로 롤백
    - JSON, YAML을 지원하고 재사용성 및 편의를 위한 수 많은 기능 포함
    - 커스텀 리소스를 사용해서 거의 모든 행동 가능 (슬랙 알림, 온프렘 데이터센터의 리소스 구성 등)

## 인프라 구성의 역사
- Pre-Cloud (긴 소요 시간)
    - 데이터 센터 확보
    - 네트워크 공사 및 운영시설 설치
    - 서버 수요 예측 및 주문
    - 서버 설치(랙)
    - 필요에 따라 라우터, 스위치 등 네트워크 기기 설치
    - OS 인스톨 및 프로그램 설치
- 클라우드 (짧은 소요 시간)
    - 필요한 리소스 선별
    - 데이터 입력
- Infrastructure as Code (매우 짧은 소요 시간)
    - 설정 파일 업로드 


### Infrastructure as Code(IAC)
- 클라우드에서 인프라를 코드로 관리하는 방법
- 장점
    - 항상 일관된 아키텍처 구현 = 휴먼 에러를 배제
    - 재사용성 = 일정 단위의 모듈로 만들어 다양한 프로젝트에서 재사용 가능, 인프라의 구성을 다른 사람과 공유하거나 복제 가능
    - 쉬운 관리 = 리소스의 변경 및 이력의 추적, 일괄 삭제 가능, Git 등을 통해 형상 관리 가능

### 실습 - 클라우드포메이션으로 EC2 웹서버 만들기
1. 서울 리전 클라우드포메이션 콘솔로 이동
2. 스택 생성
    - 템플릿 파일 업로드
    - 클라우드포메이션에서 IAM 리소스 생성 승인
    - 모든 리소스가 프로비전될 때 까지 기다리자
3. 리소스 목록 확인
    - 인스턴스 : 퍼블릭 도메인으로 접속도 해보자
4. 인스턴스 타입을 바꿔보자
    - 클라우드포메이션 스택 업데이트, 인스턴스 타입 변경(이름도 바꿔보자)
5. 리소스 정리 : 그냥 스택 삭제해버리면 됨 (편하다!)

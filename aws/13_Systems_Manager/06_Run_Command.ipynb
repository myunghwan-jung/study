# AWS SSM Run Command
- 관리중인 노드에게 원격에서 명령을 실행할 수 있는 서비스
    - 관리중인 노드 : SSM Agent가 설치된 상태에서 SSM의 관리를 받는 인스턴스/온프렘 서버
- 주로 단발성의 명령을 수행할 때 활용
    - 예 : 관리중인 서버 전체에 신규 서비스 설치, 특정 서비스 혹은 어플리케이션 재시작, 로그 파일 캡쳐 등
- 미리 준비된 문서에 지정된 명령 수행 가능
    - 문서 : AWS에서 제공하거나 유저가 직접 작성한 명령어 모음
        - 버전 관리 가능
- 태그 기반/대상 이름/인스턴스 ID 기반으로 대상 선정 : 다수의 인스턴스 클러스터에 명령 수행 가능
    - Eventual Consistency 지향 : Async 하게 명령 처리

## 데모 - 런 커맨드를 활용한 버전 업데이트
- 인스턴스를 프로비전 한 후 런 커맨드로 버전 업데이트
- 인스턴스 안의 index.html을 런 커맨드 명령으로 수정
    - 수정 시 런커맨드 내에 입력한 버전을 추가 혹은 업데이트

1. IAM 역할 생성
    - EC2 Role for SSM 선택
2. 인스턴스 2개 프로비전
    - 키페어 없음
    - 보안그룹 디폴트
    - 인스턴스 프로파일 선택
    - 유저데이터에 스크립트 입력
3. SSM 문서 생성
    - 명령 또는 세션 선택
    - 문서 유형 : command
    - 컨텐츠 타입은 JSON, 스크립트 입력
4. SSM의 인벤토리로 이동, 정보 확인
5. 생성한 문서의 명령 실행
    - 버전 입력
    - 수동으로 인스턴스 선택
6. 인스턴스의 dns로 접속해보자
    - 버저닝이 출력되는 것을 확인
7. 명령 실행 : 버전 수정
    - 인스턴스에 접속, 버저닝 확인
8. 리소스 정리 : 인스턴스

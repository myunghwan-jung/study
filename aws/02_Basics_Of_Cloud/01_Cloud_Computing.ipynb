# 클라우드 컴퓨팅이란?
- IT 리소스를 인터넷을 통해 온디맨드로 제공하고 사용한 만큼만 비용을 지불하는 것을 말한다 (AWS)
- 온디맨드 : 수요에 반응함 
- 클라우드 : 빌려 쓰기
- 클라우드 컴퓨팅 : 컴퓨팅 빌려 쓰기

## 서버-클라이언트 아키텍쳐
- 서버가 없는데 클라이언트가 많아지면? 통신이 복잡해짐.
- 클라이언트들 중간에 서버를 배치, 통신을 간단하게 만든다.
- 어플리케이션이 발전하면서 서버에 대한 수요가 증가, 이 수요에 반응하기 위해 기업들은 데이터센터를 설립.
- 데이터센터 : 어플레이케이션의 서버를 호스팅하는 실제 시설  

    - 필요 인프라  
    1. 컴퓨팅 시스템을 위한 하드웨어
    2. 네트워킹 장비
    3. 전원 공급 장치
    4. 전기 시스템
    5. 백업 발전기
    6. 환경 제어 장치 (에어컨, 냉각장치 등)
    7. 운영 인력
    8. 기타 인프라  
    - 문제점  
    1. 높은 비용
        - 건물 유지 비용, 서버 구매 비용, 유지보수 등
        - 한번 구매하면 수요에 상관없이 계속 보유
    2. 느린 구축 시간
        - 유저의 수요에 빠르게 대처하기 어렵다.
        - 장애 기기를 교체하는 시간 역시 느림.
        
- 이런 문제점들로 인해 기업들은 인프라를 클라우드로 옮기기 시작했음.

## 클라우드 컴퓨팅의 장점
- 자본 제출을 운영 지출로 대체
    - 데이터 센터 구축 비용, 서버 구매 비용 -> 운영비로 대체
    - 막대한 초기 비용 대신 쓰는 만큼 비용 지불
- 규모의 경제로 얻게 되는 이점
    - 소비 수량이 많을 수록 단가가 남음 : 규모의 경제
    - AWS의 규모의 경제로 인한 이득을 누릴 수 있음
- 용량 추정 불필요
- 속도 및 민첩성 개성
    - 몇 번의 클릭으로 바로 리소스를 확보 가능
    - 개발비용 절감
- 데이터 센터 웅영 및 유지 관리에 비용 투자 불필요
    - 인프라 관리가 아닌 비즈니스에 자원 집중 가능
- 빠른 확장성
    - 몇 번의 클릭으로 전 세계에 서비스 가능
- 용어 정리
    - 클라우드 컴퓨팅 : IT리소스를 인터넷을 통해 온디맨드로 제공하고, 사용한 만큼만 비용을 지불하는 것
    - 온디맨드 : 공급 중심이 아닌 수요에 반응해서 서비스를 제공하는 것
- 클라우드 컴퓨팅의 이점
    - 초기 투자 비용이 적음
    - 수요에 대한 빠른 대처 가능
    - 불확실한 수요 예측에서 오는 손해가 적음
    - 규모의 경제의 혜택을 볼 수 있음
    - 제품 개발에 집중 가능
    - 유지보수가 쉬움

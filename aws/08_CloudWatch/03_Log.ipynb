# CloudWatch Log
- AWS 서비스 및 온프레미스 서비스에서 수집한 로그를 저장하고 관리하고 확인할 수 있는 서비스
    - AWS의 대부분의 서비스와 기본적으로 연동 (예 : AWS Lambda 등)
- 로그 관리 및 활용을 위한 다양한 기능 제공
    - 로그 모으기 : 다양한 계정에서 하나의 계정으로 로그를 모아서 관리
    - 실시간 모니터링
    - 로그 수명 주기 관리 : 일정 기간 후에 로그를 아카이브 하거나 삭제
    - 로그 쿼리 기능 : 수집한 로그에서 일정 규칙에 따라 로그 쿼리

## 클라우드워치 로그의 구성요소
- 로그 그룹 : 로그의 관리 단위. 주로 같은 어플리케이션/서비스/리소스 단위로 분류
    - 예 : application-{스테이지}, lambda 함수 명 등
    - 다양한 설정의 기본 단위
- 로그 스트림 : 같은 소스에서 순차적으로 받는 로그들의 집합
    - 예 : EC2 클러스터로 구성된 웹 서버 중, 특정 인스턴스 아이디에서 온 로그 집합
- 로그 이벤트 : 타임스탬프와 데이터로 구성된 특정 이벤트의 기록
- 보존 기간 : 로그가 자동으로 삭제되기까지 걸리는 시간(무한정 보관도 가능)
- 로그 클래스
    - 스탠다드 : 실시간 모니터링이 필요하거나 자주 사용되는 로그(기본)
    - Infrequent Access : 자주 사용되지 않는 로그를 비용 효율적으로 관리하기 위한 클래스. 클라우드워치의 일부 기능만 활용 가능
        - 예 : Subcription Filter, Metric Filter, Insight 서비스 등 사용이 불가능
    - 로그 그룹 생성 후 변경 불가능

## Log Insight
- 대화식 검색으로 클라우드워치 로그를 분석할 수 있는 서비스
    - 제이슨 형식으로 구성된 로그를 쿼리 가능
    - 하나의 요청은 최대 20개의 로그 그룹을 동시에 쿼리 가능
- S3 -> OpenSearch 등 별도의 로그 분석 작업 필요 없이 로그 분석 가능
- 쿼리 저장 및 대시보드로 시각화 가능

## Metric Filter
- 클라우드워치 로그에 필터를 걸어 필터링된 로그를 지표화 시켜주는 기능
- 특정 필터 패턴과 일치되는 로그만 지표화
    - 다양한 비교 및 정규식 적용 가능
- 필터가 적용된 시점부터 지표화

## 로그 관련 기타 기능
- Live Tailing : 실시간으로 올라오는 로그를 확인하는 방법. 콘솔 / CLI 활용 가능
- 이상 탐지 : 머신 러닝을 활용해서 로그 그룹 단위로 패턴에서 벗어난 이상한 로그를 탐지
- Log Subscription : 실시간으로 다른 서비스/보관 장소로 로그를 전달하는 기능
    - 로그에 대한 분석, 로그 전용 계정으로 이관, 백업 등의 용도로 활용
    - 필터링 가능
    - 예 : S3에 백업, 다른 계정으로 전송, Elastic Search 등으로 전달하여 분석

### 데모
- 로컬 어플리케이션으로 클라우드워치 로그 생성
    - 로그 그룹/로그 스트림 구조 확인
- 클라우드워치 인사이트로 다양한 쿼리 동작 및 시각화까지 확인
1. IAM 사용자 생성
    - demo-my-cloudwatch-user
    - 직접 정책 연결 : cloudwatchfullaccess V2
    - 보안 자격 증명 : 액세스 키 생성, 해당 키로 스크립트 수정
    - 어플리케이션 실행
2. 클라우드워치의 로그 그룹으로 이동
    - 랜덤한 로그들이 생성된 것을 확인
    - 테일링 시작 : 로그가 올라올 때 마다 표시해줌 (어플리케이션을 재실행해보자)
3. 로그 인사이트를 통한 쿼리 사용
    - 로그 그룹 my-log-group 선택
    - 쿼리 모음집의 쿼리들을 복붙해보자!
